```python
!pip install pillow pyheif
import os
import glob
from PIL import Image
import pyheif

dataset_dir = "/content/drive/MyDrive/bancodedados_seg_linha"

# Percorre todas as subpastas e converte HEIC para JPG
for category in ["forward", "left", "right", "nothing"]:
    folder_path = os.path.join(dataset_dir, category)
    heic_files = glob.glob(os.path.join(folder_path, "*.HEIC"))  # Lista os arquivos .HEIC

    for heic_file in heic_files:
        heif_file = pyheif.read(heic_file)  # Lê o arquivo HEIC
        image = Image.frombytes(heif_file.mode, heif_file.size, heif_file.data, "raw", heif_file.mode, heif_file.stride)

        # Salvar como JPG com o mesmo nome
        jpg_path = heic_file.replace(".HEIC", ".jpg")
        image.save(jpg_path, "JPEG")
        print(f"Convertido: {heic_file} → {jpg_path}")

        # (Opcional) Remover o arquivo HEIC após conversão
        os.remove(heic_file)
```
